```python
import json
file_path = '/Users/saptarshimondal/Documents/1.Course_Materials_MSDS/2nd SEM/CSCE 670 ISR/3.Project/MFlix-RecommenderSystem/src/output/Working_poster_MovieID.json'

with open(file_path, 'r') as file:
    movie_id_lit = json.load(file)
```

```python
type(movie_id_lit)
```

```python
def check_working_url(row):
    if row['movieId'] not in movie_id_lit:
        return 'NF'
    else:
        return row['poster_working_url']
    
movies_data_new['final_poster_url'] = movies_data_new.apply(check_working_url, axis =1 )
        
```
