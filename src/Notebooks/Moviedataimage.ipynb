```python
import os

def list_image_names(folder_path):
    image_names = os.listdir(folder_path)
    return image_names
```

```python
image_folder = "Working images of posters"

image_names = list_image_names(image_folder)
```

```python
image_names
```

```python
import re

# Function to extract the numerical part from a filename
def extract_number(filename):
    # Extract digits from the filename
    match = re.search(r'\d+', filename)
    if match:
        return int(match.group())
    else:
        return None
```

```python
sorted_filenames = sorted(image_names, key=extract_number)

print(sorted_filenames)
```

```python
def filename_to_integer(filename):
    filename_without_extension = filename.replace('.jpg', '')
    return int(filename_without_extension)

# Convert each filename to integer
integer_filenames = [filename_to_integer(filename) for filename in sorted_filenames]

print(integer_filenames)
```

```python
movieID_posters = pd.DataFrame(integer_filenames, columns=['movieId'])

movieID_posters
```
