```python
IMAGE_FOLDER = '/Users/saptarshimondal/Documents/1.Course_Materials_MSDS/2nd SEM/CSCE 670 ISR/3.Project/MFlix-RecommenderSystem/src/data/IMAGES'
```

```python
data_df = pd.read_csv('/Users/saptarshimondal/Documents/1.Course_Materials_MSDS/2nd SEM/CSCE 670 ISR/3.Project/MFlix-RecommenderSystem/src/mflix-app/static/Movies_data3200.csv')
data_df
```

```python
if not os.path.exists(IMAGE_FOLDER):
    os.makedirs(IMAGE_FOLDER)
```

```python

count = 0 
incorrect = 0
for index, row in tqdm(data_df.iterrows(), total = len(data_df), desc = "Image Download"):
    url = row['poster_working_url']
    id = row['movieId']

    # Skip rows where URL is not available
    if pd.isna(url):
        continue
    
    # Get the file extension from the URL
    file_extension = url.split('.')[-1]
    
    # Define the filename for the image
    filename = f"{id}.{file_extension}"
    
    # Download the image
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code != 404:
        # Save the image
        with open(os.path.join(IMAGE_FOLDER,filename), 'wb') as f:
            f.write(response.content)
        #print(f"Image saved for movie with ID {id}")
        count+=1
    else:
        incorrect+=1 
        #print(f"Failed to download image for movie with ID {id}")

print(f"\n{count}")
```
