```python
poster_path = metadata['poster_path'].to_list()
posters = ['https://image.tmdb.org/t/p/w500' + str(path) for path in poster_path]
```

```python
import urllib3
import os
from tqdm import tqdm

final_list = []
available_posters = 0

# Initialize urllib3 PoolManager
http = urllib3.PoolManager()
#pbar = tqdm(total=len(posters), desc = 'Available posters')

for url in tqdm(posters, desc = 'Total Posters', total=len(posters)):
    try:
        # Send a GET request to download the image
        response = http.request("GET", url)
        
        # Check if the request was successful (status code 200)
        if response.status == 200:
            final_list.append(url)
            available_posters+=1
            #pbar.update(1)
                
            
        else:
            final_list.append('Image not found')
            
    except Exception as e:
        final_list.append('Image not found')

#Takes around 45 minutes to run
```

```python
available_posters
```

```python
poster_df = pd.DataFrame([final_list])
```

```python
metadata['poster_working_url'] = poster_df.T
metadata
```
